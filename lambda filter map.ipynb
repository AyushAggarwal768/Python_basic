```python
def lcm(x,*y):
    y=list(y)
    y.append(x)
    max=y[0]
    for i in y:
        if(i>max):
            max=i
        mul=1
        for i in y:
            mul=mul*i
    print(max)
    print(mul)
    for i in range(max,mul+1):
        count=0
        for j in y:
            if(i%j==0):
                count=count+1
        if(count==len(y)):
            print("lcm",i)
            break
    
lcm(13,14,50,20,10,12,15)
```
